```python
import numpy as np
import pandas as pd
```

```python
# Read file as raw text
with open("day2_input.csv") as f:
    text = f.read()
```

```python
# Split on commas
parts = [x.strip() for x in text.split(",") if x.strip()]
```

```python
# Keep only well-formed ranges
df = pd.DataFrame(parts, columns=["raw"])
```

```python
df
```

```python
def expand_range(x):
    start, end = map(int, x.split("-"))
    return list(range(start, end + 1))
```

```python
df["expanded"] = df["raw"].apply(expand_range)
```

```python
df
```

```python
df["list_length"] = df["expanded"].apply(len)
```

```python
df.head(40)
```

```python
def count_invalid_ids_in_range(start, end):
    count = 0
    min_len = len(str(start))
    max_len = len(str(end))

    for L in range(min_len, max_len + 1):
        if L % 2 != 0:
            continue  # skip odd lengths
        half_len = L // 2
        for i in range(10**(half_len-1), 10**half_len):
            candidate = int(str(i)*2)
            if start <= candidate <= end:
                count += 1
    return count
```

```python
total_invalid = 0
for product_id_range in df["expanded"]:
    start = product_id_range[0]
    end = product_id_range[-1]
    total_invalid += count_invalid_ids_in_range(start, end)

print("Total invalid IDs:", total_invalid)
```
